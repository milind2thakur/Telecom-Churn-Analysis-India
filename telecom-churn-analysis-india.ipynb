```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/kaggle/input/telecom-churn-dataset/telecom_churn.csv')

# Get unique states
states = df['state'].unique()

# Set the number of columns for the subplot grid
num_cols = 4
num_rows = (len(states) + num_cols - 1) // num_cols  # Calculate number of rows needed

# Create a figure and axes
fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, num_rows * 5), sharey=True)

# Flatten axes array for easy iteration
axes = axes.flatten()

# Plot each state in a subplot
for i, state in enumerate(states):
    ax = axes[i]
    # Filter data for the current state
    state_data = df[df['state'] == state]
    # Countplot
    sns.countplot(data=state_data, x='telecom_partner', palette='tab10', ax=ax)
    
    # Annotate each bar with count and percentage
    total_count = state_data['telecom_partner'].count()
    for p in ax.patches:
        height = p.get_height()
        width = p.get_width()
        count = int(height)
        percentage = (count / total_count) * 100
        ax.text(p.get_x() + width / 2., height + 0.5, f'{count}\n({percentage:.1f}%)', 
                ha='center', va='bottom', fontsize=10)

    ax.set_title(state)
    ax.set_xlabel('')
    ax.set_ylabel('')

# Hide unused subplots
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

# Adjust layout
plt.tight_layout()
plt.suptitle('Distribution of Telecom Partners Across States', fontsize=16, y=1.02)

# Show the plot
plt.show()
```

```python
import pandas as pd
import folium

# Load the dataset
df = pd.read_csv('/kaggle/input/telecom-churn-dataset/telecom_churn.csv')

# Aggregate data to get the average data_used per city
city_data = df.groupby('city')['data_used'].mean().reset_index()

# Example of adding latitude and longitude for each city
city_coords = pd.DataFrame({
    'city': ['Kolkata', 'Mumbai', 'Delhi', 'Chennai', 'Hyderabad', 'Bangalore'],
    'latitude': [22.5726, 19.0760, 28.6139, 13.0827, 17.3850, 12.9716],
    'longitude': [88.3639, 72.8777, 77.2090, 80.2707, 78.4867, 77.5946]
})

# Merge the city data with coordinates
city_map_data = pd.merge(city_data, city_coords, on='city')

# Create a base map
m = folium.Map(location=[20.5937, 78.9629], zoom_start=5)

# Add data to the map
for _, row in city_map_data.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=row['data_used'] / 5000,  # Adjust the radius scale as needed
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
        popup=f"City: {row['city']}<br>Average Data Used: {row['data_used']:.1f}"
    ).add_to(m)

# Save or show the map
m.save('data_used_by_city_map.html')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named df

# Convert the 'date_of_registration' column to datetime format
df['date_of_registration'] = pd.to_datetime(df['date_of_registration'])

# Sort the DataFrame by date
df = df.sort_values('date_of_registration')

# Create a cumulative count of customers by date for each telecom partner
df['customer_count'] = df.groupby('telecom_partner').cumcount() + 1

# Plotting
plt.figure(figsize=(14, 8))

for company in df['telecom_partner'].unique():
    company_data = df[df['telecom_partner'] == company]
    plt.plot(company_data['date_of_registration'], company_data['customer_count'], label=company)

plt.title('Increase in Customers Over Time for Each Telecom Partner')
plt.xlabel('Date of Registration')
plt.ylabel('Number of Customers')
plt.legend(title='Telecom Partner')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Get the latest date in the dataset
latest_date = df['date_of_registration'].max()

# Filter the data to the latest date
latest_data = df[df['date_of_registration'] == latest_date]

# Group by telecom partner and count the number of customers
customer_counts = latest_data.groupby('telecom_partner')['customer_count'].max()

# Find the telecom partner with the highest number of customers
winner = customer_counts.idxmax()
winner_count = customer_counts.max()

print(f"The telecom partner with the highest number of customers as of {latest_date.date()} is {winner} with {winner_count} customers.")
```
